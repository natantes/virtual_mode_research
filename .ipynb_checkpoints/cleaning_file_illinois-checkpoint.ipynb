```python
import pandas_access as mdb

# Define your .mdb file
import pyodbc
import pandas as pd

# Create a connection string
conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    r'DBQ=../SRC2022/SRC2022_GroupIV.mdb;'
)

# Create a connection
conn = pyodbc.connect(conn_str)

# List all tables in the database
tables = list(map(lambda t: t.table_name, conn.cursor().tables(tableType='TABLE')))

# Print all table names
print(f"All tables: {tables}")

# Initialize an empty dictionary to hold your dataframes
dataframes = {}

# Try to read each table one by one
for table in tables:
    print(f"Attempting to read table: {table}")
    try:
        df = pd.read_sql(f'SELECT * FROM {table}', conn)
        dataframes[table] = df
        print(f"Successfully read table: {table}")
    except Exception as e:
        print(f"Failed to read table: {table}")
        print(f"Error: {e}")

# Close connection
conn.close()
```
